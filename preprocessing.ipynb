```python
import pandas as pd
import numpy as np
#essential libraries for preprocessing

```

```python
df=pd.read_csv("groundwaterreport.csv")#converting the extracted csv into df 
```

```python
df
```

```python
df.replace("-",np.nan,inplace=True)#replacing all the - to Nan
```

```python
df.isnull().sum()#counting the null values
```

```python
df.info()#information about the dataframe
```

```python
columns=["S. No.","State","District","Location","Longitude","Latitude","Year","pH","EC (μS/cm)","CO3 (mg/L)","HCO3","Cl (mg/L)","F (mg/L)","SO4","NO3","PO4","Total Hardness", "Ca (mg/L)","Mg (mg/L)","Na (mg/L)","K (mg/L)","Fe (ppm)","As (ppb)","U (ppb)"]#column name from the pdf
```

```python
df.columns=columns#assigning the column names to dataframe
```

```python
df
```

```python
import string as str#converting the district,state and location names to proper format
df["State"]=df["State"].str.strip().str.title()
df["District"]=df["District"].str.strip().str.title()
df["Location"]=df["Location"].str.strip().str.title()
```

```python
df
```

```python
df.isnull().sum()#counting the null values
```

```python
df["Location"]=df["Location"].fillna("Unknown")#filling the missing location values to Unknown
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset="Latitude")#dropping 3 rows which didn't had latittudes as the main aim of our project was to plot the heatmap
```

```python
df.isnull().sum()
```

```python

df.info()
```

```python
df["pH"]=df["pH"].astype(np.float64)#type conversion of pH column
```

```python

df.info()
```

```python
columns
```

```python
cols=['EC (μS/cm)','CO3 (mg/L)','HCO3','Cl (mg/L)','F (mg/L)','SO4','NO3','PO4','Total Hardness','Ca (mg/L)','Mg (mg/L)','Na (mg/L)','K (mg/L)','Fe (ppm)','As (ppb)','U (ppb)']#columnns of which type conversion was needed to apply the mathematical opeation
```

```python
df[cols]=df[cols].apply(pd.to_numeric,errors="coerce")#type conversion errors="coerce" to make sure that Null value doesn't cause an error
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
colf=["pH",'EC (μS/cm)','CO3 (mg/L)','HCO3','Cl (mg/L)','F (mg/L)','SO4','NO3','Total Hardness','Ca (mg/L)','Mg (mg/L)','Na (mg/L)','K (mg/L)']#columns whose missing value was to be filled
```

```python
for col in colf:
    df[col] = df.groupby("District")[col].transform(    #grouping by the district and filling the median value of the columns by the median value of that district.Median was choosen above mean as it doesn't get affected by outliers
        lambda x: x.fillna(x.median())
    )
```

```python
df.isnull().sum()
```

```python
for col in colf:#there were some values whose district median was missing as all the values of all the location of the district was missing so choosed state as the next grouping 
    df[col] = df.groupby("State")[col].transform(   #grouping by the state and filling the median value of the columns by the median value of that state.Median was choosen above mean as it doesn't get affected by outliers
        lambda x: x.fillna(x.median())
    )
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer#importing knn imputer
```

```python
cols_for_knn = [
    "pH","EC (μS/cm)","CO3 (mg/L)","HCO3","Cl (mg/L)","F (mg/L)",
    "SO4","NO3","Total Hardness","Ca (mg/L)","Mg (mg/L)","Na (mg/L)","K (mg/L)",
    "PO4","U (ppb)"
]

```

```python

imputer = KNNImputer(n_neighbors=5)#applying knn imputer to the columns whose missing values were around 30-50% of the total dataset
df_imputed = imputer.fit_transform(df[cols_for_knn])#imputing using the value of all the parameter whose value were present

```

```python
df[cols_for_knn] = pd.DataFrame(df_imputed, columns=cols_for_knn, index=df.index)#converting the dataframe

```

```python
# Check missing values
df.isna().sum()
```

```python
df
```

```python
df.drop(["Fe (ppm)", "As (ppb)"], axis=1, inplace=True)#since the missing values were above 50% so had to drop these 2 columns as filling them through knn would introduce biasness in the dataset
```
